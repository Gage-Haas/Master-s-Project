```python
import os 

path = os.getcwd()

os.chdir(path +"\datasets")

print(os.getcwd())
```

```python
files = os.listdir(u'.')
files = [f for f in files if f[-3:] == 'txt']

print(files)

eliminate_set = set()
print("\n")
for i in range(0, len(files)):
    
    print("Checking for Lines To Eliminate in " + str(files[i]))
    
    line_count = len(open(files[i]).readlines(  ))
    f = open(files[i], "r")    
    line_counter = 1

    for z in range(0, line_count):
    
        currentline = f.readline()
        
        #currentline = line.split(",")
       
        if '9' in currentline :
            eliminate_set.add(line_counter)
        
        line_counter+=1
    
    f.close()

print("\nLength of Elimination Set\n")
print(len(eliminate_set))

eliminate_lines = sorted(eliminate_set)
print(type(eliminate_lines))
print("\nElimination List\n")
print(eliminate_lines)
```

```python

for i in range(0, len(files)):
    
    print("Creating Cleaned and Matched File for " + str(files[i]))
    
    line_count = len(open(files[i]).readlines(  ))
    f = open(files[i], "r")  
    
    file_name = files[i]
    file_name = os.path.splitext(file_name)[0]
    file_name = file_name + "_Cleaned_&_Matched.txt"
  
    f2 = open(file_name, "w")
    line_counter = 1

    for z in range(0, line_count):
    
        line = f.readline()
        
        if(line_counter not in eliminate_lines):
            
            f2.write(line)
        
        line_counter+=1

    f.close()
    f2.close()
```
